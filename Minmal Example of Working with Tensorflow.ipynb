# Import Libraries

```python
import tensorflow as tf
import numpy as np
observations=1000
from tensorflow.keras.models import Sequential

```

# Generate data

```python
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(low=-10,high=10,size=(observations,1))
generated_inp=np.column_stack((xs,zs))
noise=np.random.uniform(-1,1,(observations,1))
generated_op=2*xs-3*zs+5+noise
np.savez('Tf_intro',inputs=generated_inp,targets=generated_op)
```

# Train Data

```python
training_data=np.load('Tf_intro.npz')
input_size=2
output_size=1
model=tf.keras.Sequential([ tf.keras.layers.Dense(output_size)   ])
                     
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(training_data['inputs'],training_data['targets'],epochs= 100,verbose= 2)
    
    
```

# Extract Weights and Bias

```python
weights=model.layers[0].get_weights()[0]
weights
```

```python
bias=model.layers[0].get_weights()[1]
bias
```

Hence, it is clear from the observation that the weights and bias matches with the relation defined by us.

# Predicting with our model(Outputs)

```python
training_data['targets'].round(1)
```

```python
x=model.predict_on_batch(training_data['inputs'])
arr=np.array(x)
arr.round(1)
```
